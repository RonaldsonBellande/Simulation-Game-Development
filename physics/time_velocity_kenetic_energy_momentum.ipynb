```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_s = np.array(range(0, 11))
velocity_mph = np.array([0.0,0.4,0.7,0.8,1.7,1.9,2.4,3.0,3.7,4.2,4.7])
velocity_ms = velocity_mph * 0.44
```

```python
plt.plot(time_s, velocity_ms, 'o')
plt.title('Time vs. Velocity graph')
plt.xlabel("Time(s)")
plt.ylabel("Velocity(m/s)")
plt.show()
```

```python
m, b = np.polyfit(time_s, velocity_ms, 1)
plt.plot(time_s, m*time_s + b)
plt.plot(time_s, velocity_ms, 'o')
additional_info = "y = {0}x{1}".format(str(round(m,3)), str(round(b,3)))
plt.figtext(.3, .8, additional_info)
plt.title('Time vs. Velocity graph')
plt.xlabel("Time(s)")
plt.ylabel("Velocity(m/s)")
plt.show()
```

```python
k_e = (1/2)*m*velocity_ms**2
```

```python
m, b = np.polyfit(time_s, k_e, 1)
#plt.plot(time_s, m*time_s + b)
plt.plot(time_s, k_e, 'o')
model = np.poly1d(np.polyfit(time_s, k_e, 2))
k_e = np.linspace(0, 10, 5)
plt.plot(k_e, model(k_e))
additional_info = "y = {0}x{1}".format(str(round(m,3)), str(round(b,3)))
plt.figtext(.3, .8, additional_info)
plt.title('Time vs. Kinetic Energy')
plt.xlabel("Time(s)")
plt.ylabel("Kenetic Energy")
plt.show()
```

```python
p = m * velocity_ms
```

```python
m, b = np.polyfit(time_s, p, 1)
plt.plot(time_s, m*time_s + b)
plt.plot(time_s, p, 'o')
additional_info = "y = {0}x{1}".format(str(round(m,3)), str(round(b,3)))
plt.figtext(.3, .8, additional_info)
plt.title('Time vs. Momentum')
plt.xlabel("Time(s)")
plt.ylabel("Momentum")
plt.show()
```
